```python
# pdx.histogram(tx_df, 'General_2022', num_bins=10)

import matplotlib.pyplot as plt

# create a list of column names to plot
columns_to_plot = ["General_2022"]

# loop over the columns and create a histogram for each one
for col in columns_to_plot:
    # get the data for the column
    data = tx_df.select(col).rdd.flatMap(lambda x: x).collect()
    
    # create the histogram
    plt.hist(data, bins=20)
    plt.title(col)
    plt.show()
```

```python
from pyspark.sql.functions import count, when, col, desc

# Calculate the count of missing values in each column
selected_vars = ['Voters_VotingPerformanceEvenYearGeneral', 'Voters_VotingPerformanceEvenYearPrimary',
                 'Voters_VotingPerformanceEvenYearGeneralAndPrimary', 'CountyEthnic_Description', 'Religions_Description',
                 'Ethnic_Description','Parties_Description','VoterParties_Change_Changed_Party', 'Voters_Gender',
                 'Voters_Age', 'Voters_BirthDate']

tx_df.select([count(when(col(c).isNull(), c)).alias(c) for c in selected_vars]).show()
```

```python
# Show the distribution of a categorical column
tx_df.groupBy("CommercialData_EstimatedHHIncomeAmount").count().orderBy(desc("count")).show()
```

```python
# Show the top n most frequent values of a categorical column
n = 10
tx_df.groupBy("CommercialData_EstimatedHHIncomeAmount").count().orderBy(desc("count")).limit(n).show()
```
