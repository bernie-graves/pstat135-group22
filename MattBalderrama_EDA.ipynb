```python
az_df = spark.read\
    .format("csv")\
    .option("header", "true")\
    .option("nullValue", "NA")\
    .option("delimiter", "\t")\
    .option("inferSchema", "true")\
    .load(data+"VM2Uniform--AZ--2021-05-20/VM2Uniform--AZ--2021-05-20.tab")
```
